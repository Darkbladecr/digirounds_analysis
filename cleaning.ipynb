```python
import pandas as pd
```

```python
df = pd.read_csv('digirounds.csv', header=None)
df = df.T
df.columns = df.iloc[0]
df = df.drop(df.index[0])
df = df.set_index(['week', 'day'])
df = df.astype('int')
```

```python
df.info()
```

```python
df['time per patient'] = df['time'] / df['patients']
```

```python
cols = df.columns.tolist()
cols.insert(2, cols[-1])
cols = cols[:-1]
df = df[cols]
```
